```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error as mse


%matplotlib inline
```

```python
df = pd.read_csv("../csv/total_data.csv")
df = df.drop(df.columns[0], axis=1)
```

```python
fig, axs = plt.subplots(ncols=7, nrows=3, figsize=(20, 10))
index = 0
axs = axs.flatten()
for k,v in df.items():
    sns.boxplot(x=k, data=df, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

```python
fig, axs = plt.subplots(ncols=7, nrows=3, figsize=(20, 10))
index = 0
axs = axs.flatten()
for k,v in df.items():
    sns.distplot(v, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

```python
# Compute the correlation matrix
corr = df.corr()

# print(corr)
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(30, 12))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.show()
```

```python
df = df.drop(["away_team_rest_time"], axis=1)
# coz they are exactly same as can be seen from correlation matrix
```

```python
y = df["result"]
X = df.drop(["result"], axis=1)

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state = 121)
```

```python
scaler = preprocessing.MinMaxScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)

scaled_X_train = pd.DataFrame(data=scaled_X_train, columns=X_train.columns)
scaled_X_test = pd.DataFrame(data=scaled_X_test, columns=X_test.columns)

scaled_X_train
```

```python
LR = LogisticRegression()

scoring = ['accuracy', 'precision_macro', 'recall_macro' , 'f1_weighted', 'roc_auc']
scores = cross_validate(LR, scaled_X_train, y_train, scoring=scoring, cv=20)

sorted(scores.keys())
LR_accuracy = scores['test_accuracy'].mean()
LR_precision = scores['test_precision_macro'].mean()
LR_recall = scores['test_recall_macro'].mean()
LR_f1 = scores['test_f1_weighted'].mean()
LR_roc = scores['test_roc_auc'].mean()

LR.fit(scaled_X_train, y_train)
LR_pred = LR.predict(scaled_X_test)
LR_test_accuracy = accuracy_score(LR_pred, y_test)
LR_test_precision = precision_score(LR_pred, y_test)
LR_test_recall = recall_score(LR_pred, y_test)
LR_test_f1 = f1_score(LR_pred, y_test)
LR_test_roc = roc_auc_score(LR_pred, y_test)

LR_test_accuracy
```

```python
decision_tree = DecisionTreeClassifier()

scoring = ['accuracy', 'precision_macro', 'recall_macro' , 'f1_weighted', 'roc_auc']
scores = cross_validate(decision_tree, X_train, y_train, scoring=scoring, cv=20)

sorted(scores.keys())
dtree_fit_time = scores['fit_time'].mean()
dtree_score_time = scores['score_time'].mean()
dtree_accuracy = scores['test_accuracy'].mean()
dtree_precision = scores['test_precision_macro'].mean()
dtree_recall = scores['test_recall_macro'].mean()
dtree_f1 = scores['test_f1_weighted'].mean()
dtree_roc = scores['test_roc_auc'].mean()

decision_tree.fit(scaled_X_train, y_train)
dtree_pred = decision_tree.predict(scaled_X_test)
dtree_test_accuracy = accuracy_score(dtree_pred, y_test)
dtree_test_precision = precision_score(dtree_pred, y_test)
dtree_test_recall = recall_score(dtree_pred, y_test)
dtree_test_f1 = f1_score(dtree_pred, y_test)
dtree_test_roc = roc_auc_score(dtree_pred, y_test)
```

```python
SVM = SVC(probability = True)

scoring = ['accuracy','precision_macro', 'recall_macro' , 'f1_weighted', 'roc_auc']
scores = cross_validate(SVM, scaled_X_train, y_train, scoring=scoring, cv=20)

sorted(scores.keys())
SVM_fit_time = scores['fit_time'].mean()
SVM_score_time = scores['score_time'].mean()
SVM_accuracy = scores['test_accuracy'].mean()
SVM_precision = scores['test_precision_macro'].mean()
SVM_recall = scores['test_recall_macro'].mean()
SVM_f1 = scores['test_f1_weighted'].mean()
SVM_roc = scores['test_roc_auc'].mean()

SVM.fit(scaled_X_train, y_train)
SVM_pred = SVM.predict(scaled_X_test)
SVM_test_accuracy = accuracy_score(SVM_pred, y_test)
SVM_test_precision = precision_score(SVM_pred, y_test)
SVM_test_recall = recall_score(SVM_pred, y_test)
SVM_test_f1 = f1_score(SVM_pred, y_test)
SVM_test_roc = roc_auc_score(SVM_pred, y_test)
```

```python
random_forest = RandomForestClassifier()

scoring = ['accuracy', 'precision_macro', 'recall_macro' , 'f1_weighted', 'roc_auc']
scores = cross_validate(random_forest, X_train, y_train, scoring=scoring, cv=20)

sorted(scores.keys())
forest_fit_time = scores['fit_time'].mean()
forest_score_time = scores['score_time'].mean()
forest_accuracy = scores['test_accuracy'].mean()
forest_precision = scores['test_precision_macro'].mean()
forest_recall = scores['test_recall_macro'].mean()
forest_f1 = scores['test_f1_weighted'].mean()
forest_roc = scores['test_roc_auc'].mean()

random_forest.fit(scaled_X_train, y_train)
forest_pred = random_forest.predict(scaled_X_test)
forest_test_accuracy = accuracy_score(forest_pred, y_test)
forest_test_precision = precision_score(forest_pred, y_test)
forest_test_recall = recall_score(forest_pred, y_test)
forest_test_f1 = f1_score(forest_pred, y_test)
forest_test_roc = roc_auc_score(forest_pred, y_test)

forest_test_accuracy
```

```python
KNN = KNeighborsClassifier()

scoring = ['accuracy', 'precision_macro', 'recall_macro' , 'f1_weighted', 'roc_auc']
scores = cross_validate(KNN, scaled_X_train, y_train, scoring=scoring, cv=20)

sorted(scores.keys())
KNN_fit_time = scores['fit_time'].mean()
KNN_score_time = scores['score_time'].mean()
KNN_accuracy = scores['test_accuracy'].mean()
KNN_precision = scores['test_precision_macro'].mean()
KNN_recall = scores['test_recall_macro'].mean()
KNN_f1 = scores['test_f1_weighted'].mean()
KNN_roc = scores['test_roc_auc'].mean()

KNN.fit(scaled_X_train, y_train)
KNN_pred = KNN.predict(scaled_X_test)
KNN_test_accuracy = accuracy_score(KNN_pred, y_test)
KNN_test_precision = precision_score(KNN_pred, y_test)
KNN_test_recall = recall_score(KNN_pred, y_test)
KNN_test_f1 = f1_score(KNN_pred, y_test)
KNN_test_roc = roc_auc_score(KNN_pred, y_test)

KNN_test_accuracy
```

```python
models_initial = pd.DataFrame({
    'Model'       : ['Logistic Regression', 'Decision Tree', 'Support Vector Machine', 'Random Forest', 'K-Nearest Neighbors'],
    'Accuracy'    : [LR_accuracy, dtree_accuracy, SVM_accuracy, forest_accuracy, KNN_accuracy],
    'test_Acc'    : [LR_test_accuracy, dtree_test_accuracy, SVM_test_accuracy, forest_test_accuracy, KNN_test_accuracy],
    'Precision'   : [LR_precision, dtree_precision, SVM_precision, forest_precision, KNN_precision],
    'Recall'      : [LR_recall, dtree_recall, SVM_recall, forest_recall, KNN_recall],
    'F1_score'    : [LR_f1, dtree_f1, SVM_f1, forest_f1, KNN_f1],
    'AUC_ROC'     : [LR_roc, dtree_roc, SVM_roc, forest_roc, KNN_roc],
    'test_Precision'   : [LR_test_precision, dtree_test_precision, SVM_test_precision, forest_test_precision, KNN_test_precision],
    'test_Recall'      : [LR_test_recall, dtree_test_recall, SVM_test_recall, forest_test_recall, KNN_test_recall],
    'test_F1_score'    : [LR_test_f1, dtree_test_f1, SVM_test_f1, forest_test_f1, KNN_test_f1],
    'test_AUC_ROC'     : [LR_test_roc, dtree_test_roc, SVM_test_roc, forest_test_roc, KNN_test_roc],
    
    }, columns = ['Model', 'Accuracy', 'test_Acc', 'Precision', 'Recall', 'F1_score', 'AUC_ROC', 'test_Precision', 'test_Recall', 'test_F1_score', 'test_AUC_ROC'])

models_initial.sort_values(by='test_Acc', ascending=False)
```
